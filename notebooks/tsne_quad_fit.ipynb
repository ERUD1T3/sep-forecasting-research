```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.manifold import TSNE

# Generate dummy data
n_samples = 300
n_features = 50
centers = 5
random_state = 42

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=centers, random_state=random_state)

# t-SNE with different levels of early exaggeration
exaggeration_levels = [1, 12, 24]  # low, medium, high
tsne_results = []

for exaggeration in exaggeration_levels:
    tsne = TSNE(n_components=2, early_exaggeration=exaggeration, random_state=random_state)
    X_tsne = tsne.fit_transform(X)
    tsne_results.append((X_tsne, exaggeration))

# Plot the results
plt.figure(figsize=(18, 5))
for i, (X_tsne, exaggeration) in enumerate(tsne_results):
    plt.subplot(1, 3, i + 1)
    scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=15)
    plt.title(f't-SNE with Early Exaggeration = {exaggeration}')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.colorbar(scatter, ax=plt.gca())

plt.suptitle('t-SNE Visualization with Different Levels of Early Exaggeration')
plt.show()
```
